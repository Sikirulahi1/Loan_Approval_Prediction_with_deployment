# Understanding the Dataset features

- no_of_dependents: This refers to the number of dependents (family members, children, etc.) that the applicant has.

- education: This likely indicates the education level of the applicant. It could be a categorical variable with values like "Graduate" and "Not Graduate."

- self_employed: This could be a binary variable indicating whether the applicant is self-employed or not.

- income_annum: This represents the annual income of the applicant.

- loan_amount: This is the amount of money the applicant is applying to borrow.

- loan_term: This indicates the term or duration of the loan in terms of months.

- cibil_score: CIBIL score (Credit Information Bureau (India) Limited score) is a credit score used in India to assess an individual's creditworthiness. It's a numerical representation of an individual's credit history.

- residential_assets_value: This might refer to the value of residential assets (such as a house) owned by the applicant.

- commercial_assets_value: Similarly to residential assets, this might indicate the value of commercial assets owned by the applicant.

- luxury_assets_value: This could be the value of luxury assets owned by the applicant, such as high-end vehicles or other valuable possessions.

- bank_asset_value: This might refer to the value of assets held by the applicant in banks or financial institutions.

- loan_status: This is likely the target variable that you're trying to predict. It indicates the status of the loan application, which could be "Approved" or "Rejected."

# Removing trailing spaces

```python
df = df.drop(['loan_id'], axis = 1)
df
```

```python
df.columns.tolist()
```

```python
def remove_trailing_spaces(dataset):
    dict = {}
    old_column = [col for col in dataset.columns]
    new_column = [col.strip() for col in dataset.columns]
    for old, new in zip(old_column, new_column):
        dict[old] = new
    dataset.rename(columns=dict,inplace=True)
    return dataset
```

```python
df = remove_trailing_spaces(df)
```

# Pipeline for the data preprocessing

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as imbpipeline
from imblearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedShuffleSplit
from imblearn.over_sampling import RandomOverSampler,SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import cross_validate
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

```python
def cross_validate_model(dataset, model,names = feature_names,scaler = MinMaxScaler(), sampler = SMOTE(), scoring = "roc_auc"):
    X = dataset[names]
    Y = dataset.loan_status
    x_cols = X.columns
    for features in x_cols:
        X[features] = scaler.fit_transform(X[[features]])
    
    splitter = StratifiedShuffleSplit(n_splits=5,random_state = 42,test_size=0.2)
    for train, test in splitter.split(X, Y):
        x_train, y_train = X.loc[train], Y.loc[train]
        x_test, y_test = X.loc[test], Y.loc[test]
        
    undersample = RandomUnderSampler()
    
    # oversample = SMOTE()
    steps = [("o", sampler), ("u", undersample)]
    pipeline = Pipeline(steps = steps)
    x, y = pipeline.fit_resample(x_train, y_train)
    score = cross_validate(model, x, y, scoring = "roc_auc")
    # print("F1 Cv score:",score)
    return score['test_score']
```

```python
log = LogisticRegression()
cross_validate_model(df,log)
```

```python
rfc = RandomForestClassifier()
cross_validate_model(df, rfc)
```

# Lets train our model

```python
def model_training(dataset, model,feature_names = feature_names, scaler = MinMaxScaler()):
    X = dataset[feature_names]
    Y = dataset.loan_status
    
    x_cols = X.columns
    
    for col in feature_names:
        X[col] = scaler.fit_transform(X[[col]])
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size= 0.2, random_state = 0)
    
    undersample = RandomUnderSampler()
    oversample = SMOTE()
    steps = [("o", oversample), ("u", undersample)]
    pipeline = Pipeline(steps = steps)
    x, y = pipeline.fit_resample(x_train, y_train)
    model.fit(x, y)
    preds = model.predict(x_test)
    score = model.score(x_test, y_test)
    
    return model, preds
```
